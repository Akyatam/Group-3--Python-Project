# Blacks, Asians, and Hispanics were more adversely affected than Whites by the pandemic-induced recession

In 2020, employment fell drastically for all race and ethnicity groups, as seen by declines in the employment–population ratios for Whites, Blacks, Asians, and Hispanics.
- There was an improvement in the second half of the year though it was not substantial enough to make up for the fall that occurred in the second quarter. 
- If we compare the ratios of 4th quarters of 2020 with 4th quarter of 2019, it can be observed that some groups were affected more than others. Although the ratio for Whites decreased by 3.2 percent points over the year, to 57.9 percent, the declines in the employment–population ratios for Blacks, Hispanics, and Asians were much more
- The ratio for Blacks decreased to 53.9 percent in the fourth quarter of 2020, a loss of 5.1 percentage points over the year. 
- The employment–population ratios for Hispanics and Asians also fell sharply during 2020, with the ratio for Hispanics decreasing by 4.8 percentage points, to 59.6 percent, and the ratio for Asians decreasing by 4.4 percentage points, to 58.2 percent


## Analysing COVID Impact by deriving Unemployment Rate Across All Races per quarter of 2020 and comparing results of 4th quarter of 2020 from 2019 4th quarter

### Finding Unemployment Rate- We have total Labor Force and unemployed population count (Level in Thousands) across all races. To derive at unemployment rate we need to divide total labor force with Unemployed count

### Slicing Required Rows from Main Dataset and Reseting Index

```python
df6 = Employment_status.filter(items = [0,4,5,9,10,14,15,19],axis=0)
df6_reset=df6.set_index('Characteristic')
df6_reset
```

### Using Transpose to divide Labor force by No.of.Unemployed People Race wise

```python
df7=df6_reset.transpose()
df7
```

```python
df8=df7.iloc[0:5]
df8
```

### Dividing column to derive at unemployment rate

```python
import warnings
warnings.filterwarnings('ignore')
df8['White'] = df8['White Unemployed']/df8['White Civilian labor force']*100
df8['Black'] = df8['Black or African American Unemployed']/df8['Black or African American Civilian labor force']*100
df8['Asian'] = df8['Asian Unemployed']/df8['Asian Civilian labor force']*100
df8['Hispanic or Latino ethnicity'] = df8['Hispanic or Latino ethnicity Unemployed']/df8['Hispanic or Latino ethnicity Civilian labor force']*100
df8
```

### Slicing Required Columns

```python
df9=df8.iloc[:,[8,9,10,11]]
df9
```

```python
df9.plot()
plt.title('Unemployment Rate Across Races', size = 20,color="g")
plt.xlabel('Year',size=20)
plt.ylabel('Unemployment Rate', size = 20)
plt.xticks(size = 10)
plt.yticks(size = 15)
plt.legend(loc="upper right",fontsize=10)





plt.show()
```

## Analysis

- Among the major race and ethnicity groups, unemployment rates at the end of 2020 were much higher than in the fourth quarter of 2019.
-  As evident from the chart, in 2nd quarter of 2020 unemployment rate have increased tremendously for all the races.
- The unemployment rate for Blacks, at 10.3 percent in the fourth quarter of 2020, increased by 4.5 percentage points over the year. 
- The jobless rate for Asians more than doubled, increasing by 4.0 percentage points over the year, to 6.7 percent. -The rate for Hispanics increased by 4.7 percentage points, to 8.9 percent. 
- The unemployment rate for Whites, at 6.0 percent, increased by 2.8 percentage points over the year.
- Therefore it can be concluded that Blacks were affected the most followed by Hispanic and Asians.
