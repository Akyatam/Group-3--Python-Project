```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Wage_Gap.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(df.columns[[1, 3, 4, 7]], axis=1)
df
```

```python
salaried = df.loc[(df['LOCATION'] == "USA") & (df['SUBJECT'] == "EMPLOYEE") & (df['TIME'] >= 2010)]
salaried #only analyzing wage gap between salaried men & women from 2010-2020
```

```python
salaried.plot(x ='TIME', y='Value', kind = 'bar')
plt.show()
```

```python
average_salaried = salaried["Value"].mean()
max_salaried = salaried["Value"].max()
min_salaried = salaried["Value"].min()
print("The average wage gap in the last 10 years between salaried men & women in the US is " + str(average_salaried) + " percent")
print("The minimum wage gap in the last 10 years between salaried men & women in the US is " + str(min_salaried) + " percent")
print("The maximum wage gap in the last 10 years between salaried men & women in the US is " + str(max_salaried) + " percent")
```
